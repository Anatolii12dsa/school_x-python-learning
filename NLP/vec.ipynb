```python
morph = pymorphy3.MorphAnalyzer()

def words_to_pos_tags(word_list):
    result = []
    for word in word_list:
        
        parsed_word = morph.parse(word)[0]
        
        pos_tag = parsed_word.tag.POS
        #
        result.append(f"{word}_{pos_tag}")
    return result
```

```python
word_res = []

for i in range(len(array_list)):
    a = array_list[i]
    res = -1  
    max_word_index = -1  
    for j in range(len(array_list)):
        if j == i:
            continue
        else:
            b = array_list[j]
            cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
            
            if cos_sim > res:
                max_word_index = j
                res = cos_sim
    word_res.append(max_word_index)

print(word_res)

```
