```python
import torch
import torch.nn as nn # Модуль PyTorch для создания нейронных сетей.
import torch.optim as optim  # Модуль PyTorch для оптимизаторов.

```

```python
X = torch.tensor([[-1, -1], [-1, 1], [1, -1], [1, 1]], dtype=torch.float32)
Y = torch.tensor([0, 1, 1, 0], dtype=torch.float32)
```

```python
class XORNet(nn.Module):
    def __init__(self):
        super(XORNet, self).__init__()
        self.fc1 = nn.Linear(2, 4)  
        self.fc2 = nn.Linear(4, 1)  

    def forward(self, x):
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x
```

```python
net = XORNet()
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.1)
```

```python
for epoch in range(100000):  # Проходим по данным 100000 раз
    optimizer.zero_grad()  # Обнуляем градиенты
    outputs = net(X)  # Прямое распространение
    loss = criterion(outputs, Y.view(-1, 1))  # Вычисляем функцию потерь
    loss.backward()  # Обратное распространение
    optimizer.step()  # Обновление весов

    if epoch % 1000 == 0:
        print(f'Эпоха {epoch}, потеря: {loss.item()}')
```

```python
with torch.no_grad():
    test_data = torch.tensor([[-1, -1], [-1, 1], [1, -1], [1, 1]], dtype=torch.float32)
    predictions = net(test_data)
    for i, prediction in enumerate(predictions):
        print(f'Вход: {test_data[i].tolist()}, Выход: {round(prediction.item())}')
```
