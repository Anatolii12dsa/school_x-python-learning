Позволяет использовать регулярные выражения для токенизации, что полезно для настройки токенизации под конкретные потребности. Например, можно указать шаблон, который будет соответствовать отдельным словам в тексте. При вызове метода tokenize на экземпляре токенизатора будет произведено разбиение текста на токены в соответствии с указанным шаблоном

 Легкий токенизатор, который хорошо справляется с многими европейскими языками.
 Простота и скорость: ToktokTokenizer разработан с акцентом на производительность и эффективность. Он быстро разделяет текст на токены, что делает его хорошим выбором для обработки больших объемов текста.

Разбиение на слова: Токенизатор разбивает текст на слова, основываясь на пробелах и других пробельных символах. Он старается сохранить целостность слов, не разделяя их на составные части, если это необходимо.

Поддержка для разных языков: ToktokTokenizer спроектирован так, чтобы поддерживать несколько языков. Это полезно, если вы работаете с многоязычными текстами.

Сохранение знаков препинания: По умолчанию, токенизатор сохраняет знаки препинания в токенах. Это может быть важным для некоторых задач анализа текста.

Инкапсулирован в NLTK: ToktokTokenizer включен в состав библиотеки NLTK, поэтому его можно использовать в сочетании с другими инструментами и ресурсами NLTK.

```python
toktok = nltk.tokenize.ToktokTokenizer()
print(toktok.tokenize(book))
```


TreebankWordTokenizer - это один из токенизаторов в библиотеке NLTK, который использует набор правил, схожих с правилами Penn Treebank, для разделения текста на слова. Этот токенизатор широко применяется в обработке текста и лингвистических исследованиях. Вот некоторые характеристики и особенности TreebankWordTokenizer:

Разбиение на слова: TreebankWordTokenizer разбивает текст на слова, основываясь на пробелах, знаках препинания и других символах. Он старается сохранить целостность слов, не разбивая их на составные части, если это необходимо.

Учёт знаков препинания: Этот токенизатор сохраняет знаки препинания как отдельные токены, что может быть полезно для некоторых задач анализа текста, таких как выделение пунктуации.

Совместимость с Penn Treebank: TreebankWordTokenizer следует стандартам Penn Treebank, что делает его полезным для сравнения и анализа текстов, которые следуют этим стандартам.

Инкапсулирован в NLTK: Этот токенизатор включен в состав библиотеки NLTK и может быть легко использован в сочетании с другими инструментами и ресурсами NLTK.

Обработка различных текстов: TreebankWordTokenizer подходит для обработки различных видов текста, включая новости, научные статьи, литературные произведения и другие жанры.

```python
treebank = nltk.tokenize.TreebankWordTokenizer()
print(treebank.tokenize(book))
```

Токенизатор, который использует регулярные выражения для разделения текста на слова и знаки препинания.

```python
wordpunct = nltk.tokenize.WordPunctTokenizer()
print(wordpunct.tokenize(book))
```

```python
word = nltk.tokenize.NLTKWordTokenizer()
tokens = word.tokenize(book)

# wordpunct = nltk.tokenize.WordPunctTokenizer()
# tokens = wordpunct.tokenize(book)

total_tokens = len(tokens)
unique_tokens = len(set(tokens))
lexical_diversity = unique_tokens / total_tokens 
print(f"Лексическое разнообразие книги: {lexical_diversity * 100}%")
```

```python
class SimpleTokenizer:
    def tokenize(self, text):
        words = text.split()  # Разделение текста по пробелам
        return words


tokenizer = SimpleTokenizer()
tokens = tokenizer.tokenize(book)
print(tokens)
```
